위의 데이터는 다양한 방법으로 **'표현'** 될 수 있습니다. 몇 가지 다른 방법으로 위 데이터를 표현해 보죠.

![](https://d3s0tskafalll9.cloudfront.net/media/images/F-10-5.max-800x600.jpg)

총 네 가지 방법으로 왼쪽의 데이터를 표현해 보겠습니다.

데이터는 **분자 형태**로 표현될 수 있습니다. 아니, 사실 분자 형태는 실제 데이터(물체) 그 자체라고도 할 수 있죠. 아무런 가공을 거치지 않은, 데이터가 존재하는 형태 그대로의 표현입니다.

혹은 **이미지**로 표현해 볼 수도 있습니다. 분자 형태, 혹은 데이터 그 자체보다는 아주 조금 가공된 형태입니다. '시각적' 표현이라고도 할 수 있겠습니다. 보통 이미지를 컴퓨터에서 표현할 때에는 (14x14)와 같이 특정 사이즈를 가지는 2차원 배열로 표현됩니다. 컬러 이미지라면 R, G, B의 3채널을 가지기 때문에 3차원 배열로 표현하기도 하죠. 디지털 세상에서 RGB 색상은 이 [링크](https://www.w3schools.com/colors/colors_rgb.asp)처럼 Red, Green, Blue 세 가지의 색의 조합으로 나타납니다. 색의 3원색을 각각 어느 정도로 섞느냐에 따라 무지갯빛 색상을 모두 만들어낼 수 있죠. 따라서 컬러 이미지는 각 픽셀 포인트마다 Red, Green, Blue, 총 세 가지의 값을 가지므로 `(14x14)` 크기의 2차원 배열을 Red 용, Green 용, Blue 용 총 세 장 사용해 `(14x14x3)`의 3차원 배열로 표현합니다. 물론 흑백 이미지라면 RGB가 필요없이 흑과 백의 밝은 정도를 0~255 사이의 수치로만 표현하면 되므로 `(14x14)` 한 장, 즉 2차원 배열로 표현됩니다.

세 번째 표현 방법은 **표**로 나타내는 것입니다. 해당 데이터가 가지고 있는 '특징(Feature)'을 각 열에 나타내고, 해당 열마다 어떤 값을 가지는지를 표 형식으로 나타낼 수 있죠. 다만, 표 형식으로 나타낼 때에는 먼저 해당 데이터를 '어떤 특징(Feature)으로 나타낼 것인지', 즉 각 열의 내용을 적합한 것들로 정의하는 과정이 필요합니다. 위와 같은 데이터를 표로 나타내려고 할 때 '손가락의 길이'와 같은 열이 있으면 안 되겠죠. 즉, 데이터를 표로 나타내기 위해서는 사람이 개입해야 합니다.

마지막 표현 방법은 **카테고리**입니다. 카테고리는 머신러닝에서 주로 '예측하고자 하는 값'으로 사용되는 것이기도 합니다. 데이터를 카테고리로 표현하려면 '어떤 카테고리 공간이 있는지'와 '그중 어떤 카테고리에 속하는지'를 정하는 과정이 필요합니다. 즉, 카테고리 공간이 `['무궁화', '수국', '진달래', '카네이션']`이라면 그중 '무궁화'라는 알맞은 카테고리로 표현할 수 있습니다. 하지만 만약 카테고리 공간이 `['판다', '호랑이', '거북', '강아지']`로 되어 있다면 그 안에서 아무리 잘 선택해봐야 '판다' 혹은 '호랑이' 정도로 밖에 표현할 수 없습니다. 즉, 카테고리를 활용한 표현은 사람의 개입이 매우 강하게 들어갑니다. 사람이 카테고리 공간을 잘 정의해놔야, 해당 데이터가 알맞은 카테고리로 표현될 수 있습니다.

네, 위와 같이 네 가지 방법으로 데이터를 표현해 보았습니다. 이게 대체 무슨 의미가 있길래 이렇게 설명하는 걸까요?

위의 네 가지 표현 방법은 언뜻 보기에는 아무런 관련이 없어 보일 수 있지만, 사실 아주 강력한 **계층적(Hierarchical)** 인 관계를 갖고 있습니다.
무슨 말인고 하니, 1번부터 4번으로 갈수록 **사람의 개입**이 강하게 들어가는 표현이라는 겁니다.

다음과 같이 나타낼 수 있죠.

![](https://d3s0tskafalll9.cloudfront.net/media/images/F-10-6.max-800x600.jpg)

위에 있는 분자 형태와 이미지의 표현은 `Raw`한 `Sensory`한 표현이라고 할 수 있습니다. 특별한 가공이 들어가지 않은 날것의, 그저 감각기관만으로 받아들일 수 있는 데이터이죠.

반면, 표와 카테고리의 아래쪽 표현들은 `Abstract`하고 `Internal`한 표현들입니다. 사람의 개입이 강하게 들어가고, 그렇기 때문에 더 추상적이고 함축적인 내재적 표현을 할 수 있게 되죠.

만약 입력되는 이미지의 크기가 `(224x224x3)`이었다면 약 15만 개의 숫자로 데이터가 표현됩니다. 이를 잘 학습된 딥러닝 네트워크에 먹이면 단 1,000개의 숫자만으로 그 이미지의 내재된 특성들을 모두 표현할 수 있게 되는 것입니다. 놀랍게도 사람이 가르쳐준 규칙 없이도요! (물론 네트워크를 학습시키는 것은 사람의 몫입니다. 하지만 한 번 잘 학습시켜 놓으면 몇 천, 몇 만 장의 이미지를 수 분 내에 처리할 수 있다는 것이 중요한 포인트죠)

데이터의 '내재된 표현(Internal Representation)'이라는 것이 무엇인지, 그리고 데이터로부터 '좋은 표현(Good Representation)'을 추출해낸다는 것의 의미는 무엇인지 조금 와닿으시나요?


## Q5. 데이터의 내재된 표현이란 무엇이고, 데이터로부터 좋은 표현을 추출해낸다는 것은 무슨 뜻인가요? 그리고 이는 어떻게 수행될 수 있나요?

- 데이터의 내재된 표현이란, 그 데이터가 담고 있는 정보의 총체, 혹은 함의를 나타내는 표현을 말한다. 다만, raw한 sensory 데이터는 그 자체로는 중요한 의미를 담지 못하고 있을 수 있다. 마치 이미지를 멀리서 보면 강아지가 보이지만, 각 픽셀 하나하나의 값들은 강아지와 직접적인 연관이 없는 것과 같다.
- 이러한 맥락에서 데이터로부터 표현을 추출해낸다는 것은 데이터 안에 내재되어 있는, 추상적인 표현을 추출한다는 것을 의미한다. 강아지가 찍힌 이미지에서 좋은 표현을 추출해내려면 픽셀 하나하나의 값으로부터 '강아지'라는 의미를 담고 있는 벡터를 추출해낼 수 있어야 할 것이다.
- 딥러닝의 본질은 이렇게 데이터로부터 내재된 표현을 추출해내는 것이며, 이는 모델을 Gradient Descent 기반의 학습으로 수행할 수 있다. 물론 더 뛰어난 모델을 얻기 위해서는 모델의 구조나 최적화 기법 등의 세부적인 요인들을 연구하는 것이 필수적이다.

이러한 맥락에 의해 딥러닝을 'Representation Learning'이라고 칭하기도 합니다. **데이터의 표현을 학습한다는 것**, 특히 사람이 가공하지 않은, 이미지나 자연어 같은 '날 것의' 데이터를 입력하면 알아서 **그 안에 내재된 표현을 추출한다는 것**, 이것이 바로 딥러닝이 추구하는 목표의 본질입니다!

## Q6. 행동주의란 무엇인가요? 행동주의는 인간의 내면이 선천적이라고 주장하나요, 아니면 후천적이라고 주장하나요?

- 행동주의는 심리학을 과학적으로 접근하고자 하는 시도를 가진 분야로, 철저하게 경험적이고 실증적으로 인간의 내면을 접근하고자 하였다.
- 따라서 무의식과 같이 관찰될 수 없는 것을 배제하였으며, 인간의 행동은 자극으로부터 직접적으로 만들어진다고 주장하였다.
- '자극 → 행동'의 흐름에 따라, 인간의 지능 및 내면은 살아가면서 받는 자극으로 형성되는 후천적인 것이라고 주장하였다.

행동주의는 인간, 혹은 인간뿐만 아니라 모든 생명체의 지능은 반복되는 자극에 따라 형성된다고 주장하였습니다.
행동주의를 연구한 대표적인 학자 스키너(Skinner)는 비둘기에게 적절한 보상을 줌으로써 탁구를 학습시키는 것이 가능하다는 것을 실험으로 보여주었죠.

## Q7. 스키너의 상자가 무엇인가요? 스키너의 상자를 통해 어떤 실험이 진행되었나요?

- '스키너 상자(Skinner Box)'는 동물행동을 연구하기 위해 고안된 장치로, 쥐가 먹이를 얻기 위해 누를 수 있는 레버나 막대가 달려 있고, 통풍과 방음설비가 된 상자로 구성된다.
- 레버를 누르면 먹이를 받을 수 있는데, 쥐는 처음에는 이러한 사실을 알지 못하지만 레버를 누르고 먹이를 받는 경험을 여러 번 하게 될 경우 이를 학습하게 되며, 먹이를 받기 위해 레버를 누르는 행동을 할 수 있게 됨을 보였다.
- 이렇게 어떤 행동을 한 뒤에 생명체가 원하는 것을 제공하는 것을 강화(Reinforcement)라고 한다.

이렇게 행동주의는 생명체가 학습하는 과정을 단순하게 자극과 행동의 반복으로 이루어지는 것이라고 설명합니다.

이 중 생명체가 자신에게 유리한 결과를 가져다주는 행동을 알게 된다면 그 행동의 빈도를 높인다는 '강화이론'은, 머신러닝의 방법론 중 '강화학습(Reinforcement Learning)'의 근간이 되는 이론이기도 합니다. 강화학습은 실제로 특정 행위자를 학습시키기 위해 행위자가 하는 행동이 얼마나 좋았는지에 대한 척도인 '보상'을 부여하고, 행위자는 그 보상을 높이기 위해 자신의 행동을 교정하면서 학습해나가게 되죠.

## Q8. 인지주의는 행동주의와 어떻게 다르며, 어떠한 새로운 관점을 제시하였나요?

- 인지주의는 '자극 → 행동'의 관계만 설명하던 행동주의에 대해 반발하며 '자극 → (정보처리) → 행동' 의 관점을 제시하였다.
- 인간의 행동이 단순히 자극으로부터 만들어진다는 주장으로 인간의 '의식'을 부정했던 행동주의와 달리, 인지주의는 인간이 자극을 받은 후 내면에서 정보를 처리하고 가공하는 의식이 존재함을 주장하였다.
- 즉, 인지심리학은 인간의 마음을 일종의 정보 처리 체계로 보고 접근하고, 이에 따라 인간의 뇌가 정보를 처리하는 과정에 대해 관심을 가지며 신경과학이라는 새로운 분야를 만들어냈다.

인지주의는 자극과 반응 사이에 있는 '정보처리' 과정에 관심을 두었습니다. 생명체의 지능은 단순히 어떤 행동을 자극에 맞추어 '강화'하는 것을 넘어서서, 자극을 받아 반응을 하기까지 복잡한 알고리즘을 통해 정보를 처리하는 것이라고 생각한 거죠. 다음과 같은 그림으로 나타낼 수 있습니다.

## Q9. 연결주의에 따르면 지능체는 어떻게 학습을 진행해 나가나요?

- 연결주의는 자극으로부터 반응을 하기까지의 과정에서 뉴런과 같이 연결되어 있는 모형이 정보를 처리해 나간다고 설명한다.

조금 더 엄밀하게 고민해 보죠.

우리는 고등학생 때까지 일차, 이차, 삼차 함수와 같은 다항 함수뿐만 아니라, 지수 함수와 로그함수, 삼각함수 등 어마무시하게 다양한 함수들을 배웠습니다.
이 모든 것들의 상위 집합인 '함수'는 수학적으로 다음과 같이 표현됩니다.

y = f(x)
y=f(x)
여기에서 xx는 입력되는 값, 그리고 yy는 출력되는 값으로, 조금 전에 봤던 이미지처럼 xx가 ff라는 박스에 들어가 yy라는 결과물로 출력된다는 것을 식으로 나타낸 것이죠.

반면, 우리가 지금 배우고 있는 프로그래밍에도 함수의 개념이 있습니다. 파이썬으로 함수를 구현하자면 다음과 같죠.

```python
def f(x):
    y = 3 * x + 2
    return y
```

위에서 구현된 함수 f(x)는 f(x) = 3x+2f(x)=3x+2 라는 아주 간단한 일차 함수를 구현한 것입니다.

다음과 같이 사용하면 수학에서의 표기와 동일한 형태로 프로그래밍에서의 함수를 사용할 수 있습니다.

```python
x = 5

y = f(x)   # 3 * 5 + 2 = 17
y
```

여기서 한번, $f(x) = x^2 + e^{x+1}$ 를 파이썬 함수로 구현하고, $f(5)$ 값을 계산해 보세요. $(f(5)=428.43$ 정도의 값이 나와야 합니다.)<br>
(hint. exp 함수는 파이썬에 내장된 math 라이브러리 안에 있습니다.)

```python
import math

def f(x):
    y = x ** 2 + math.exp(x+1)
    return y

f(5)
```

### Q15. 스칼라와 벡터를 간단히 설명해 보세요.

- 스칼라는 '크기'만 가지고 '방향'은 가지지 않는 양이다. 주로 단 하나의 숫자로 표현한다.
- 벡터는 '크기'와 '방향'을 두 가지 가지는 양이다. 주로 여러 개의 숫자로 표현되며, 좌표 상에서 길이와 방향을 가지는 화살표로 표현한다.

이러한 배경에서, 함수는 아래 세 가지 방법을 통해 입력을 출력으로 매핑할 수 있습니다.


**✓ One-to-One**

One-to-one 매핑은 입력값 $x$와 출력값 $y$가 모두 스칼라인 경우입니다. 단 하나의 값을 입력했을 때, 단 하나의 값만 나오게 됩니다.

![](https://d3s0tskafalll9.cloudfront.net/media/images/F-10-17.max-800x600.jpg)

One-to-one의 예제를 살펴봅시다.

![](https://d3s0tskafalll9.cloudfront.net/media/original_images/f-10-table1.png)

위와 같이 키 값이 있을 때, 몸무게를 예측하는 함수를 만든다면 One-to-One에 해당합니다.
키 값을 입력받으면 일정한 연산을 통해 몸무게를 출력하는 함수를 만들어볼 수 있겠죠. 다음과 같은 그림으로 나타낼 수 있습니다.

![](https://d3s0tskafalll9.cloudfront.net/media/images/F-10-18.max-800x600.jpg)

$f$를 아주 간단한 선형 방정식으로 채택한다면, 다음과 같은 식으로 쓸 수 있습니다.

$$f(x) = wx + b = \hat y$$

하지만 사람의 키와 몸무게는 단순한 일차함수의 관계를 갖지 않습니다. 몸무게는 키뿐만 아니라 보다 복잡한 요인들로 인해 결정되겠죠.
그러므로 위와 같은 단순한 일차 함수는 키 값을 입력받았을 때 정답값이 아닌, 예측값을 출력할 것입니다. 이러한 예측값은 $y$ 위에 모자(hat)를 씌워서 $\hat y$로 표기합니다.

# 17-11. 신경망의 본질 (5) 함수와 모델

**우리가 배운 함수와, 앞으로 다룰 '모델'이라는 함수가 다른 점**


## 함수와 모델의 차이점
---
함수에 대한 이야기를 마치기 전에 한 가지만 더 짚고 넘어가겠습니다.

지금까지 우리가 알고 있는 친숙한 형태의 함수와, 앞으로 받아들일 '신경망', 혹은 신경망 외의 '머신러닝 모델' 등 형태의 함수가 어떻게 다른지 생각해 보죠.
(머신러닝/딥러닝에서 데이터를 입력받아 원하는 값을 예측하거나, 원하는 형태로 데이터를 표현하여 출력하는 모든 함수를 '모델'이라고 칭합니다.)

우리가 지금까지 다뤄왔던 함수는 그 형태가 고정되어 있는 경우가 대부분이었습니다.
우리가 지금까지 수학 문제를 풀든, 프로그래밍을 하며 함수를 구현하든 우리가 사용하고자 하는 함수는 정해져 있고, 우리가 해야 할 일은 그 함수의 수식을 찾아내거나, 또는 알맞은 형태로 구현하는 것뿐이었죠.

예를 들어 고등학생 때 공부하던 수학을 떠올려봅시다.

우리가 두 점 (1, 2), (0, 5)를 지나는 이차 함수를 구하는 문제를 푼다면 $f(x) = ax^2 + bx + c$와 같이 함수식을 미지수와 함께 두고 점을 대입하여 풀면 될 겁니다.

이 경우 저 두 점을 지나는 정확한 함수가 정해져 있기 때문에 $a$와 $b$, $c$도 단 한 가지 값으로 정해집니다.

하지만 우리가 머신러닝, 혹은 딥러닝에서 다루어야 하는 함수는 정확히 단 하나로 정해져있는 함수도 아닐뿐더러, 그 함수가 이차 함수인지, 삼차 함수인지, 혹은 신경망 형태로 나타날지조차 알지 못합니다. 우리에게 주어진 것이라곤 단 하나! **데이터** 뿐이죠.

(여기서 잠깐, 신경망이라는 함수는 일차 함수일까요, 이차 함수일까요, 아니면 어떤 형태의 함수일까요? 이 물음에 대한 답은 바로 다음에 다룹니다. 그전에 한번 생각해 보세요!)

집값을 맞추는 문제를 풀 때 우리에게는 집의 평수, 위도와 경도, 층수 등 각 집의 특징을 나타내는 데이터가 있을 겁니다. 이러한 특징 데이터를 함수(모델)에 입력했을 때 우리는 모델이 집의 가격을 정확히 맞추기를 바라죠. 하지만 현실 세계에서 집의 가격은 아주 복잡한 요인들이 관여합니다 (아마 현 시기의 경제 상황이라던가, 그런 설명하기 복잡한 것들까지 반영되었을 겁니다!). 그런 모든 요인을 정량화하여 함수에 넣어줄 수 없다면 우리는 완벽한 함수는 만들어낼 수 없을 것입니다.

그렇기에 머신러닝과 딥러닝은 '완벽한 함수'를 찾는 수학 문제를 푸는 것이 아니라, '그나마 가장 잘 근사할 수 있는 함수'에 조금씩 가까워지도록 시도하는 것에 가깝습니다.

![](https://d3s0tskafalll9.cloudfront.net/media/images/F-10-23.max-800x600.jpg)


## 그나마 나은 함수를 찾기 위해
---
정리해 보겠습니다. 우리가 '그나마 가장 잘 근사할 수 있는 함수'를 찾기 위해 해야 할 일은 두 가지입니다.


**1. 먼저, 모델을 어떤 함수로 나타낼 것인지를 정해야 합니다.**

- 입력 데이터가 하나인 경우, 즉 One-to-One 문제의 경우에는 $f(x)=wx+b$와 같은 일차 함수를 사용할 수 있습니다.
- 입력 데이터가 여러 개의 특징으로 이루어진 벡터인 경우, 즉 Many-to-One 문제의 경우에는 $f(x) = wx_1 + wx_2 + ... + wx_n$과 같은 다변수 선형 함수로 정할 수 있죠.
- 물론 신경망 모델을 사용할 수도 있습니다.

다양한 모델(함수) 중 어떤 함수를 채택하느냐는 모델의 최종 성능에 굉장히 중요한 영향을 미칩니다.

만약 간단한 선형 함수만으로 데이터를 잘 표현할 수 있다고 판단된다면 선형함수를 사용해도 되지만, 이미지나 자연어같이 형태가 고정되지 않은 '비정형 데이터'같은 경우 최근에는 거의 필연적으로 신경망을 사용합니다. 신경망의 구조는 여러 계층이 쌓인 형태이며 이렇게 'Hierarchial(계층적인)' 구조가 비정형 데이터의 표현을 단계별로 잘 추출해낼 수 있다는 것이 실험적으로 보여졌기 때문입니다.

이렇게 어떤 데이터에 대한 문제를 풀기 위해 '이러한 형태의 함수가 유리할 것'이라고 판단하고 함수 공간을 정하는 것을 **Inductive Bias**, 혹은 **Prior**를 가정한다고 합니다.

**Inductive Bias**란 **데이터를 설명할 수 있는 최적의 함수가 특정한 함수 공간에 존재할 것이라는 가설**을 말합니다.

예를 들어, 우리가 함수 공간을 $f(x) = wx+b$로 일차 함수 형태로 정해놓고 시작한다면 우리는 데이터를 더 잘 표현할 수 있는 $f(x) = x^2 + 5x + 2$와 같은 이차함수는 영영 못 찾게 될 겁니다. 혹은, 이미지나 자연어와 같은 데이터를 처리할 때에 신경망의 형태로 모델을 정해놓고 시작하는 것은 신경망이 비정형 데이터를 표현하는 데에 강력하다는 실험적 결과에 따른 가설에 입각하는 것이죠.

이러한 Inductive Bias가 중요한 이유는 우리가 학습시킬 모델이 세상의 모든 데이터를 볼 수 없기 때문입니다. 모델이 학습하는 과정에서 보지 못한 데이터에 대해서도 예측을 잘 하려면(즉, 일반화(Generalization)가 잘 되려면), 일반적인 패턴을 잘 반영할 수 있는 형태의 모델이어야 합니다. 그렇기에 어떤 문제를 풀고자 할 때 알맞은 함수 형태로 모델을 선택하는 것은 그 데이터에 대한 도메인 지식을 가지고 있는 사람의 몫이며, 더 좋은 모델을 찾기 위해서는 그 데이터에 알맞은 함수 공간, 즉 Inductive Bias를 잘 설정하는 것이 중요합니다.

**2. 모델의 함수 형태를 정했다면, 해당 함수 공간 안에서 최적의 함수를 찾아나갑니다.**

모델을 어떤 함수 형태로 사용할 것인지 정했다면, 그 후로는 그 함수 공간 안에서 찾을 수 있는 최적의 함수로 학습을 해 나갑니다.

이 단계가 보통 머신러닝/딥러닝에서 말하는 '모델 학습'에 해당하죠.
머신러닝 모델을 '학습'시킨다는 것은, 처음에는 데이터를 입력받아 전혀 엉뚱한 값을 내뱉던 모델을, 점점 정답에 해당하는 값을 출력할 수 있도록 '보정'해나가는 과정이라고도 할 수 있습니다.

모델을 학습시키는 방법은 모델 형태에 따라 다양합니다. 딥러닝의 신경망 모델은 'Gradient Descent(경사하강법)'이라는 방법을 사용하죠.


### Q19. 데이터를 잘 표현/예측할 수 있는 모델(함수)를 찾아내기 위한 두 가지 단계는 무엇인가요?

- 첫 번째는, 모델을 어떤 함수 형태로 나타낼 것인지 함수 공간을 정하는 단계입니다. 선형 함수를 쓸지, 의사결정 나무(Decision Tree) 모델을 쓸지, 혹은 신경망 모델을 쓸지 정해야 하며, 이 과정에서 Inductive Bias가 발생합니다.
- 두 번째, 함수 공간을 정했다면 그 안에서 최적의 함수를 찾아 학습합니다. 신경망의 경우 경사하강법과 같은 방법으로 점차 최적의 함수로 학습될 수 있습니다.


### Q20. Inductive Bias란 무엇인가요?

- Inductive Bias란, 데이터를 설명할 수 있는 최적의 함수가 특정한 함수 공간에 존재할 것이라는 가설을 말합니다.
- 모든 데이터는 그 데이터가 무엇이냐에 따라 데이터를 잘 표현/예측할 수 있는 함수의 형태가 다릅니다. 어떤 데이터는 간단한 선형 함수만으로도, 어떤 데이터는 아주 복잡한 신경망으로 나타내야만 표현이 가능할 수도 있죠. 데이터에 대한 도메인 지식을 활용하면 데이터를 가장 잘 표현할 수 있는 함수는 어떤 형태일 것이라는 가설을 세울 수 있습니다. 더 좋은 모델을 만들기 위해서는 좋은 Inductive Bias가 필요합니다.

# 17-12. AI의 현재와 미래 (1) 머신러닝과 딥러닝

**현시대 AI: 어디까지 왔나, 그리고 어디를 향해 가는가**

여러분은 어떠셨나요? 그 과정이 흥미롭고 즐거우셨길 바랍니다.

마지막으로 지금까지 알아본 내용을 갈무리하면서, 현재의 AI는 어디까지 왔는지, 그리고 어디를 향하고 있는지 알아보겠습니다.

지금까지 알아본 철학에 기반하였을 때, 머신러닝과 딥러닝이 어떤 차이를 가지는지, 그리고 두 가지가 추구하는 것이 어떻게 다르다고 설명할 수 있을까요?


### Q21. 머신러닝과 딥러닝은 어떻게 다른가요? 이제 여러분의 언어로 설명해 보세요.

머신러닝은 주어진 데이터로부터 패턴을 학습해서 원하는 값을 예측하는 것에 방점이 있다면, 딥러닝은 입력된 데이터에 내재되어 있는 표현 그 자체를 나타내도록 학습하는 것이 목표입니다. 다음 이미지로 그 차이가 명확하게 드러나죠.

![](https://d3s0tskafalll9.cloudfront.net/media/images/F-10-24.max-800x600.jpg)
[https://bigdatapath.wordpress.com/2019/09/13/ai-vs-machine-learning-vs-deep-learning-whats-the-difference/](https://bigdatapath.wordpress.com/2019/09/13/ai-vs-machine-learning-vs-deep-learning-whats-the-difference/)

기존의 머신러닝은 데이터를 입력하기 위해 사람이 직접 피처(Feature)를 가공해 주어야 했습니다. 전통적인 머신러닝 모델들은 주로 '표' 형태로 된 정형 데이터를 처리하도록 설계되었기 때문입니다. 그래서 지금까지 이미 해봤거나 앞으로도 그럴 예정이듯, 머신러닝 모델을 학습시킬 때에는 데이터를 전처리하고 모델에 들어가도록 이쁘게 가공하는 작업이 필수적입니다. 또한, 주어진 데이터를 이리저리 조합해서 더욱 좋은 피처를 만들어내는 것도 성능을 높이기 위한 굉장히 중요한 과정이죠. 이러한 과정을 통틀어 피처 엔지니어링(Feature Engineering) 이라고 합니다. 피처 엔지니어링은 그 자체로 머신러닝 내에서 한 챕터를 차지할 만큼 중요하고 방대한 내용이 많은 분야입니다. 그만큼 머신러닝은 더 좋은 모델을 만들기 위해 '사람의 개입'을 굉장히 많이 필요로 합니다.

한편, 딥러닝은 모델의 복잡성을 늘려 사람이 개입하는 피처 엔지니어링을 최소화하고자 합니다. 이미지라면 픽셀 형태로 된 2차원 배열을 그대로 신경망에 입력하면, 알아서 그 안에 있는 특징이나 표현을 학습하도록 하죠. 물론 자연어라면 한글이나 알파벳을 넣을 순 없기에 자연어를 숫자로 이루어진 벡터 형태로 가공하는 작업은 해야 하지만(자연어 임베딩(Embedding)) 그럼에도 머신러닝과는 조금 다릅니다. 머신러닝으로는 자연어를 입력받아 기계 번역을 한다던가, 문장을 생성하는 등의 태스크를 수행할 수 없죠. 이처럼 딥러닝은 뚜렷한 형태가 없는(비정형) 데이터로부터 표현을 추출해 내는 것에 매우 강합니다.

딥러닝에서는 이렇게 데이터를 모델에 입력해서 바로 원하는 출력값을 얻어내는 End-to-End Learning을 추구하는 경우가 많습니다. 끝에서 끝까지 모두 담당한다는 정도의 의미가 되겠네요.

### Q22. 글에서 설명하는 아이의 학습 과정은 무엇인가요?

- 사물을 보고 그것이 뭔지 알아가는 과정에서 ‘나무’의 다양한 예시들을 더 많이 접하게 되며, 머릿속에서는 맨 처음 봤던 나무의 예시인 ‘갈색의 딱딱하고… 매달려 있는 대상’만이 존재하였던 상태에서, 나무의 예시들을 확장적으로 받아들이면서 동시에 ‘나무’라는 대상의 핵심이 되는 특질을 서서히 압축해 나간다.

### Q23. 글에서는 선형모델이 무엇이라고 설명하나요?

직선(또는 평면, 초평면 등)의 형태로 정의되는 러닝 모델을 특별히 선형 모델(linear model)이라고 부르며, 이는 가장 단순한 러닝 모델에 해당한다.


### Q24. 오늘날 이미지 인식 분야에서 선형모델을 쓰지 않는 이유는 무엇이며, 그에 대한 대안으로 떠오른 것은 무엇인가요?

- 선형 모델은 가장 단순하고 (선형 모델로 학습이 가능한 데이터 집합에 한해서) 학습 속도가 빠르며 인간이 이해하기 쉬운 대신, 원 데이터 자체가 수많은 픽셀로 이루어져 있는 이미지와 같은 데이터의 경우에는 데이터의 복잡성을 선형 모델에 충분히 반영하는 것이 불가능하다.
- 그래서, 실제 이미지 인식 분야에서는 선형 모델 대신 인공신경망(artificial neural networks)과 같은 복잡성이 높은 러닝 모델을 사용한다. 이렇게 복잡성이 높은 신경망(neural networks) 구조를 러닝 모델로 사용하는 머신러닝 패러다임을 ‘딥러닝(deep learning)‘이라고 한다.

### Q25. 퍼셉트론이란 무엇인가요?

- 입력 벡터 $x=(x_1,x_2,...,x_d)$를 받아들인 뒤 각 성분에 가중치를 곱하는 선형 모델을 거쳐, 그 결과를 모두 합산한 후 활성화함수 $σ(⋅)$을 적용한 함수로 이루어진 모델이다. 전체 과정을 수행하는 이 선형 모델을 하나의 함수 $h(⋅)$로 나타낼 수 있으며, 이를 퍼셉트론(perceptron)이라고 부른다.
- 즉, 퍼셉트론은 선형 함수와 그 결과를 비선형 함수 활성화 함수까지 거치는 합성 함수라고 할 수 있다.


### Q26. 다층 퍼셉트론이란 무엇인가요?

- 퍼셉트론은 세 개의 층으로 이루어진다. 입력 벡터가 자리잡는 층을 입력층(input layer), 최종 출력값이 자리잡는 층을 출력층(output layer), 입력층과 출력층 사이에 위치하는 모든 층을 은닉층(hidden layer)이라고 한다.
- 이 때 퍼셉트론을 기본 빌딩 블록으로 하여, 이런 패턴에 따라 2차원적으로 연결되어 구성되는 인공신경망의 일종을 특별히 다층 퍼셉트론(MLP: multi-layer perceptron)이라고 한다.


### Q27. 심층 신경망이란 무엇인가요?

- 다층 퍼셉트론에서 은닉층의 개수가 많아질수록 인공신경망이 ‘깊어졌다(deep)’고 부르며, 이렇게 충분히 깊어진 인공신경망을 러닝 모델로 사용하는 머신러닝 패러다임을 바로 딥러닝(Deep Learning)이라고 한다.
- 이 때 딥러닝을 위해 사용하는 충분히 깊은 인공신경망을 심층 신경망(DNN: Deep neural network)이라고 한다.

- 완전 연결 신경망, 컨볼루션 신경망, 순환 신경망이다.

### Q29. 글에서 소개하는 딥러닝의 강점은 무엇인가요?

딥러닝은 ‘요인 표현 학습(feature representation learning)’ 능력을 가지고 있으며, 이는 원본 데이터로부터 최적의 성능을 발휘하는 데 사용될 수 있는 요인 표현 방법을 스스로 학습하고, 이를 기반으로 최적의 성능을 발휘하는 가중치를 더욱 효과적으로 찾는 능력을 말한다.


### Q30. 컨볼루션 신경망의 낮은 층(입력에 가까운 층)에서는 어떤 특징을, 높은 층(출력에 가까운 층)에서는 어떤 특징을 추출하나요?

- 컨볼루션 신경망의 낮은 층에 해당할수록, 이미지 상의 각 세부 영역에서의 경계, 명암, 색상 변화 등 저수준(low-level)의 특징들을 요인 표현으로 포착하는 경향을 보인다. 이렇게 위치 별 저수준 요인은, 컨볼루션 신경망의 높은 층으로 올라가면서 서로 가까운 것들끼리 조합되고, 좀 더 넓은 영역에서 고수준(high-level)의 특징들을 요인 표현으로 포착하는 경향을 보인다.
- 이렇게 낮은 층에서는 점, 선, 면과 같은 간단한 요인들을 추출하고, 점점 높은 층에 다다를수록 복잡한 요인들을 추출하며 최종적으로 이미지를 인식해나간다.


### Q31. 글에서 소개하는 딥러닝의 약점은 무엇인가요?

딥러닝 모델이 효과적으로 학습하기 위해서는 굉장히 많은 데이터가 필요하며, 학습 연산량이 매우 높아 학습 시간이 오래 걸린다는 것이 약점이다.

### Q32. 딥러닝이 블랙박스가 아니라고 주장하는 이유는 무엇인가요?

- 신경망은 여러 개의 함수를 엮어 놓은 것이라고 할 수 있고, 그 함수는 본질적으로 회귀 식과 형태가 같다.
- 즉, 여러 함수의 연산은 간단한 계산으로 이루어져 있고, 그 함수들이 합성된 형태로 수가 어떻게 연산되는지, 어떤 원리로 동작하는지 알기 때문에 블랙박스라고 할 수 없다.

**'왜 그런 결과를 내놓았는지 설명하지 못하므로 블랙박스다' 라는 주장**

- [못 믿겠다 AI, 설계자도 심층신경망 작동 방식 몰라](https://www.joongang.co.kr/article/22394025)


### Q33. 기사에서는 왜 딥러닝이 블랙박스라고 이야기하나요?

- 딥러닝의 구조는 매우 복잡하고 그 연산량 또한 엄청나게 많기 때문에 그 계산 과정을 전체적으로 이해하는 것은 불가능하다.
- 학습된 딥러닝 모델이 어떤 결과를 내놓을지 예측할 수 없으며, 모델이 내놓은 결과가 어떤 판단 근거에 따라 판단되었는지 또한 설명할 수 없다.
- 예를 들어 어느 은행의 인공지능 프로그램이 컴퓨터가 당신의 대출을 거부한 경우, 어떤 근거로 거부 결정을 내렸는지는 실제로 프로그램을 만든 사람조차도 알 수 없다. 이러한 성질로 인해 딥러닝은 블랙박스라고 할 수 있다.

어떤가요? 신경망이 동작하는 원리를 알고 있으니 블랙박스라고 할 수 없다는 주장과, 모델이 특정 결과를 내놓았을 때 어떤 점에 기인하여 그 결과를 내었는지 설명할 수 없으므로 블랙박스라고 이야기하는 설명이 있었습니다.

두 주장 모두 일리가 있습니다. 하지만 보통 비즈니스에 딥러닝 기술을 접목하는 경우에는 모델이 만들어 낸 결과가 '왜 그렇게 나왔는지'를 설명할 수 있어야 하는 경우가 많습니다. 인공지능이 만들어내는 결과를 믿기 위해 이루어지는 이러한 연구를 Interpretable Machine Learning, 혹은 Explainable AI (XAI, 설명 가능한 인공지능) 라고 합니다. 관련된 포스팅을 또 읽어보겠습니다.

- [설명가능 인공지능이란?](https://realblack0.github.io/2020/04/27/explainable-ai.html)


### Q34. 머신러닝 모델에 대해 해석을 할 수 있어야 하는 이유는 무엇인가요? 혹은 그로 인한 이점은 무엇인가요?

- 모델이 예측한 결과가 어떠한 과정으로 도출되었는지 설명할 수 있다면 납득할 만한 과정을 거쳐 도출되었다는 것을 확인함으로써 그 결과를 신뢰할 수 있다.
- 상용화 및 대중화를 염두에 두고 머신러닝 기술이 핵심이 되는 제품을 출시하고자 할수록, 제품 사용자가 제품을 신뢰하기 위한 설명의 중요성은 더 높아진다.

이렇게 머신러닝/딥러닝 모델은 예측하거나 출력한 결과가 어떻게 도출되었는지 설명할 수 있어야 신뢰도를 높일 수 있습니다.

# 17-14. AI의 현재와 미래 (3) 현시대 AI

**어디까지 왔나, 그리고 어디를 향해 가는가**

오늘 과정은 현재 AI가 어디까지 왔는지, 그리고 앞으로 해결하고자 하는 문제는 무엇인지에 대해 알아보고 마무리할 예정입니다.

2020년 6월 초에 OpenAI에서 발표한 모델 GPT-3를 필두로 거대 모델이 등장한 이후 딥러닝은 놀랄만한 결과물을 보여줍니다.

GPT-3 API를 사용하면 위키피디아의 'Bread' 페이지에서 'why is bread so fluffy?'라는 질문을 검색했을 경우 마치 구글에서 검색을 하듯 그 질문의 답이 들어있는 문단을 바로 찾아줍니다. 질문의 의도와 문맥을 이해하고, 페이지 내에 있는 내용 또한 이해해서 질문에 맞는 부분을 찾아주는 거죠.

거의 사람에게 물어봤을 때 해당 부분을 찾아주는 것과 비슷한 태스크라고도 할 수 있을 정도입니다.

이번에는 리눅스 쉘의 코드를 입력해 주는 시나리오입니다. 사람이 코드를 직접 짜는 대신, 그냥 자연어로 질문합니다. 'what day is it?' 하면, 리눅스에서 오늘 날짜를 출력해 주는 명령어인 `date +%A` 라는 명령어를 생성해냅니다. 더 재미있는 건, 오타를 내기도 합니다. 요일과 날짜를 함께 출력하기 위해서는 `date +%A %F`가 아니라 `date '+%A %F'`와 같이 따옴표를 입력해야 하는데, 이렇게 오타가 나서 사람이 'I think you forgot the quotes'라고 알려주자 따옴표(quote)를 올바르게 입력한 명령어를 다시 잘 만들어냅니다.
